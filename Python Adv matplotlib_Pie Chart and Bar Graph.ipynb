```python
Pie chart and Bar graphs making using of matplotlib.
```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# Basic Approach:
```

```python
y = np.array([35,25,25,15])
```

```python
plt.pie(y)
plt.show()
```

```python
# Req: To add the labels
```

```python
mylabels = ['Apples','Bananas','Cherries','Dates']
```

```python
plt.pie(y, labels = mylabels)
plt.show()
```

```python
# How to control the angles on the pie chart....default angle will start at 0.
```

```python
plt.pie(y, labels = mylabels, startangle = 90)
plt.show()
```

```python
# Explode...
```

```python
myexplode = [0.2,0,0,0]
```

```python
plt.pie(y, labels = mylabels, explode = myexplode)
plt.show()
```

```python
# Shadow...
```

```python
plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show()
```

```python
# Print the legends...
```

```python
plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.legend()
plt.show()
```

```python
plt.pie(y, labels = mylabels)
plt.legend()
plt.show()
```

```python
Creating bars with matplotlib:
```

```python
x = np.array(["A","B","C","D"])
y = np.array([3,8,1,10])
```

```python
plt.bar(x,y)                      # Vertical axes
plt.show()
```

```python
plt.barh(x,y)                     
plt.show()
```

```python
# Custom color:
```

```python
plt.bar(x,y, color = 'red')
plt.show()
```

```python
plt.bar(x,y, color = 'hotpink')
plt.show()
```

```python
# Width control:
```

```python
plt.bar(x,y, width = 0.1)
plt.show()
```
